```python
import os
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt

df = pd.read_csv('Data/Champ/champ.csv')
df.describe()
```

```python
df.head()
```

```python
# Rename the last feature to represent the game ID
# Check other features for any glaring name problems

df = df.rename(columns={"New" : "game_id"})
list(df)
```

```python
# Replace spaces in features with underscores for simplicity

for name in df.columns:
    df = df.rename(columns={name : name.replace(' ', '_')})
```

```python
# Remove all games not consisting of 3 players on each team

df['game_id'] = df.game_id.astype('str')
df = df.groupby('game_id')
df = df.filter(lambda x: len(x) == 6)
```

```python
df.info()
```

```python
# This section deals with team stats

'''
Create opponent columns for each entry for machine learning training.
This allows the model to check how a team stacks up to its oppoenet 
instead of to the population of all teams across all games.
'''

# Create dataframe for team stats

_ = df.to_dict()
df_stat = pd.DataFrame(columns=_)
players_grouped_by_game = df.groupby(['game_id', 'team'], as_index=False)
df_stat = players_grouped_by_game.mean()
# df_stat['game_goal_avg'] = df.groupby(['game_id']).mean()
df_stat.head(6)
```

```python
# add win column

df_stat['game_goal_avg'] = df_stat['goals'].groupby(df_stat['game_id']).transform('mean')

df_stat['win'] = df_stat[['goals', 'game_goal_avg']].apply(lambda x: 1 if x['goals'] > x['game_goal_avg'] else 0, axis=1)
df_stat.head()
```
