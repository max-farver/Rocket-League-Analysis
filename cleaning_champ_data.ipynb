```python
import os
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot

pd.options.display.max_columns = None
```

```python
df = pd.read_csv('Data/Champ/champ.csv')
df.describe()
```

```python
df.head()
```

```python
# Rename the last feature to represent the game ID
# Check other features for any glaring name problems

df = df.rename(columns={"New" : "game_id"})
list(df)
```

```python
# Replace spaces in features with underscores for simplicity

for name in df.columns:
    df = df.rename(columns={name : name.replace(' ', '_')})
```

```python
# Remove all games not consisting of 3 players on each team

df['game_id'] = df.game_id.astype('str')
df = df.groupby('game_id')
df = df.filter(lambda x: len(x) == 6)
```

```python
# This section deals with team stats

'''
Create opponent columns for each entry for machine learning training.
This allows the model to check how a team stacks up to its oppoenet 
instead of to the population of all teams across all games.
'''

# Create dataframe for team stats

_ = df.to_dict()
df_stat = pd.DataFrame(columns=_)
players_grouped_by_game = df.groupby(['game_id', 'team'], as_index=False)
df_stat = players_grouped_by_game.mean()

df_stat.info()
```

```python
# Create and fill new columns

num_rows = len(df_stat)
num_rows
fields = list(df_stat)[2:]
for field in fields:
    new_field = str(field) + '_diff'
    df_stat[new_field] = 0.00

    
for i in range(1, num_rows,2):
    for field in fields:
        new_field = str(field) + "_diff"
        
        blue_val = df_stat.at[i-1, field]
        
        orange_val = df_stat.at[i, field]
        
        orange_new_val = orange_val - blue_val
        

        blue_new_val = blue_val - orange_val
        
        
#         if field == 'goals':
#             print(orange_new_val, blue_new_val, orange_new_val + blue_new_val)
#         df_stat.set_value(i-1, new_field, orange_new_val)
        df_stat.at[i-1, new_field] = float(blue_new_val)
#         if field == 'goals':
#             print(blue_new_val, df_stat.at[i-1, new_field])
        df_stat.at[i, new_field] = float(orange_new_val)
#         df_stat.set_value(i, new_field, blue_new_val)

df_stat.head()
```

```python
# Create a column signifying a WIN

df_stat['win'] = df_stat.apply(lambda x: 1 if x['goals_diff'] > 0 else 0, axis=1)
df_stat.head()
```

```python
# Save as a quicker to access datatype HDF

df_stat.to_hdf('champ_clean_not_scaled.h5', key='df_stat')
```
